# Summary

### Heart Rate block :
<ol>
    <li>QRSAmpn's -> class 2</li>
    <li>Rn's -> class 0</li> 
    <li>All heart rates blocks after normalization -> class 3</li>
</ol>

### Frequency block :
All columns -> all classes are divided

### Histogram block :
<ol>
    <li>Bin1 + Bin2 -> all classes are divided</li>
    <li>Bin6 + Bin7 + Bin8 -> all classes divided</li>
</ol>

### Hence given a row based on the above features from each block we can detect the class

# Rules

```python
# Bin6 + 7 + 8: 

# >=0.03 && < 0.1 -> 0
# == 0 -> 2
# == 0.06 -> 3
# >= 0.1 -> 4

# Bin 1 + 2

# >= 0.61 && <= 0.7 -> 0
# > 0.71 && <= 1.0 -> 2
# == 0.67 -> 3
# >= 0.44 && <= 0.57 -> 4

# Mpfns

# >= 85 && <= 94 -> 0
# >= 5 && <= 23 -> 2 (91 -> outlier)
# >= 46 && <= 55 -> 4
# cant tell anything about 3

# Rn

# >= 23 && <= 37 -> 0
# cant tell Anything about other classes

# QRSAmpns

# >= 0.21 && <= 0.57 -> 2
# cant tell Anything about other classes

# All Heart Rate Block colums except SOI
# >= 6.172936 && <= 6.2308927 -> 3

row_index = 5 # Example row index
columns1 = ['Bin1', 'Bin2']  # Specify the column names
columns2 = ['Bin6', 'Bin7', 'Bin8']
columns3 = ['Mpfn', 'Mpfn-1', 'Mpfn-2', 'Mpfn-3', 'Mpfn-4', 'Mpfn-5', 'Mpfn-6']
columns4 = ['Rn', 'Rn-1', 'Rn-2']
columns5 = ['QRSAmpn', 'QRSAmpn-1', 'QRSAmpn-2']
columns6 = ['nHRn', 'nHRn-1', 'nHRn-2', 'nQRSAmpn', 'nQRSAmpn-1', 'nQRSAmpn-2', 'nRn', 'nRn-1', 'nRn-2']

# Sum the values of the specified columns in the given row
bin1_2 = df.loc[row_index, columns1].sum()
bin6_7_8 = df.loc[row_index, columns2].sum()
mpfns = df.loc[row_index, columns3].sum()
rns = df.loc[row_index, columns4].sum()
qrsampns = df.loc[row_index, columns5].sum()
heart_rate_block = df.loc[row_index, columns6].sum()

print(bin1_2)
print(bin6_7_8)
print(mpfns)
print(rns)
print(qrsampns)
print(heart_rate_block)

# Histogram block
if(bin1_2 >= 0.61 and bin1_2 < 0.7) :
    print("class 0(bin1_2)")

if(bin6_7_8 >= 0.03 and bin6_7_8 < 0.1) :
    print("class 0(bin6_7_8)")

if(bin1_2 > 0.71 and bin1_2 <= 1.0) :
    print("class 2(bin1_2)")

if(bin6_7_8 == 0) :
    print("class 2(bin6_7_8)")

if(bin1_2 >= 0.44 and bin1_2 <= 0.57) :
    print("class 4(bin1_2)")

if(bin6_7_8 >= 0.1) :
    print("class 4(bin6_7_8)")


# frequency Block
if(mpfns >= 85 and mpfns < 95) :
    print("class 0(mpfns)")

if(mpfns >= 5 and mpfns <= 23) :
    print("class 2(mpnfs)")

if(mpfns >= 46 and mpfns <= 55) :
    print("class 4(mpfns)")
    

# Heart rate block
if(rns >= 23 and rns <= 37) :
    print("class 0(rns)")
    
if(qrsampns >= 0.21 and qrsampns <= 0.57) :
    print("class 2(qrsampns)")

if(heart_rate_block >= 6.172936 and heart_rate_block <= 6.2308960 ) :
    print("class 3(hrb)")
```

# Rules - Updated

```python
# Rn 
# >=19 && <= 37 -> 0

#Qrsampn
# >= 0.21 && <= 0.57 -> 2

# Bin 1 2
# >= 0.59 and <= 0.70 -> 0
# >= 0.19 and <= 0.20 -> 1

# Bin 3 4 5
# >= 0.22 && <= 0.32 -> 0
# >= 0.2 && <= 0.21 -> 1

# Hrn rn qrsampn - norm 
# >= 1.33 && <= 1.35 -> 4

#hrn rn qrsampn -2 - norm
# >= 1.61 && <= 1.68 -> 4

# hrn rn qrsampn -1 -norm 
# >= 0.901393 && <= 0.969540 || >= 1.44 && <= 1.48 -> 3

# all mpfns
# >= 85 && <= 100 -> 0
# >= 46 && <= 55 -> 4

#diff -> extra/new features (Bin 3, 4, 5), [(Hrn -1+ rn -1 + qrsampn-1) -> 3]

class_values = []  # List to store the values of the "Class" column
actual = 0
print("actual_value", end = '\t')
print("predicted_value")
for index, row in df.iterrows():
    class_value = row["Class"]  # Access the value of the "Class" column for the current row
    class_values.append(class_value)

for i in range(1,26,1):
    
    actual_value = class_values[i - 1]
    cl = [0,0,0,0,0]

    columns1 = ['Bin1', 'Bin2']  # Specify the column names
    columns2 = ['Bin3', 'Bin4', 'Bin5']
    columns3 = ['Mpfn', 'Mpfn-1', 'Mpfn-2', 'Mpfn-3', 'Mpfn-4', 'Mpfn-5', 'Mpfn-6']
    columns4 = ['Rn', 'Rn-1', 'Rn-2']
    columns5 = ['QRSAmpn', 'QRSAmpn-1', 'QRSAmpn-2']
    columns6 = ['nHRn-1', 'nQRSAmpn-1', 'nRn-1']
    columns7 = ['nHRn-2', 'nQRSAmpn-2', 'nRn-2']
    columns8 = ['nHRn', 'nQRSAmpn', 'nRn']
    
    row_index = i
    
    # Sum the values of the specified columns in the given row
    bin1_2 = df.loc[row_index, columns1].sum()
    bin3_4_5 = df.loc[row_index, columns2].sum()
    mpfns = df.loc[row_index, columns3].sum()
    rns = df.loc[row_index, columns4].sum()
    qrsampns = df.loc[row_index, columns5].sum()
    h_r_q_1 = df.loc[row_index, columns6].sum()
    h_r_q = df.loc[row_index, columns7].sum()
    h_r_q_2 = df.loc[row_index, columns8].sum()

    if(rns >= 19 and rns <= 37) :
        cl[0] += 1 
    if(bin1_2 >= 0.59 and bin1_2 <= 0.70) :
        cl[0] += 1
    if(bin3_4_5 >= 0.22 and bin3_4_5 <= 0.32) :
        cl[0] += 1
    if(mpfns >= 85 and mpfns <= 100):
        cl[0] += 1

    if(bin1_2 >= 0.19 and bin1_2 <= 0.20):
        cl[1] += 1
    if(bin3_4_5 >= 0.2 and bin3_4_5 <= 0.21):
        cl[1] += 1

    if(qrsampns >= 0.21 and qrsampns <= 0.57):
        cl[2] += 1

    if( (h_r_q_1 >= 0.901393 and h_r_q_1 <= 0.969540) or (h_r_q_1 >= 1.44 and h_r_q_1 <= 1.48) ):
        cl[3] += 1

    if(mpfns >= 46 and mpfns <= 55 ) :
        cl[4] += 1
    if(h_r_q_2 >= 1.61 and h_r_q_2 <= 1.68 ):
        cl[4] += 1
    if(h_r_q >= 1.33 and h_r_q <= 1.35):
        cl[4] += 1

    predicted_value = -1
    mx = 0
    for i in range(5):
        if(mx <= cl[i]):
            mx = cl[i]
            predicted_value = i

    print(actual_value, end = "\t\t")
    print(predicted_value)

    if(actual_value == predicted_value):
        actual += 1

print("accuracy : ", end = " ")
print( float(actual) / float(27) )
```
