# Summary

### Heart Rate block :
<ol>
    <li>QRSAmpn's -> class 2</li>
    <li>Rn's -> class 0</li> 
</ol>

### Frequency block :
All columns -> all classes are divided

### Histogram block :
<ol>
    <li>Bin1 + Bin2 -> all classes are divided</li>
    <li>Bin6 + Bin7 + Bin8 -> all classes divided</li>
</ol>

### Hence given a row based on the above features from each block we can detect the class

# Rules

```python
# Bin6 + 7 + 8: 

# >=0.03 && < 0.1 -> 0
# == 0 -> 2
# == 0.06 -> 3
# >= 0.1 -> 4

# Bin 1 + 2

# >= 0.61 && <= 0.7 -> 0
# > 0.71 && <= 1.0 -> 2
# == 0.67 -> 3
# >= 0.44 && <= 0.57 -> 4

# Mpfns

# >= 85 && <= 94 -> 0
# >= 5 && <= 23 -> 2 (91 -> outlier)
# >= 46 && <= 55 -> 4
# cant tell anything about 3

# Rn

# >= 23 && <= 37 -> 0
# cant tell Anything about other classes

# QRSAmpns

# >= 0.21 && <= 0.57 -> 2
# cant tell Anything about other classes

# All Heart Rate Block colums except SOI
# >= 6.172936 && <= 6.2308927 -> 3

row_index = 15  # Example row index
columns1 = ['Bin1', 'Bin2']  # Specify the column names
columns2 = ['Bin6', 'Bin7', 'Bin8']
columns3 = ['Mpfn', 'Mpfn-1', 'Mpfn-2', 'Mpfn-3', 'Mpfn-4', 'Mpfn-5', 'Mpfn-6']
columns4 = ['Rn', 'Rn-1', 'Rn-2']
columns5 = ['QRSAmpn', 'QRSAmpn-1', 'QRSAmpn-2']
columns6 = ['nHRn', 'nHRn-1', 'nHRn-2', 'nQRSAmpn', 'nQRSAmpn-1', 'nQRSAmpn-2', 'nRn', 'nRn-1', 'nRn-2']

# Sum the values of the specified columns in the given row
bin1_2 = df.loc[row_index, columns1].sum()
bin6_7_8 = df.loc[row_index, columns2].sum()
mpfns = df.loc[row_index, columns3].sum()
rns = df.loc[row_index, columns4].sum()
qrsampns = df.loc[row_index, columns5].sum()
heart_rate_block = df.loc[row_index, columns6].sum()

print(bin1_2)
print(bin6_7_8)
print(mpfns)
print(rns)
print(qrsampns)
print(heart_rate_block)

# Histogram block
if(bin1_2 >= 0.61 and bin1_2 < 0.7) :
    print("class 0(bin1_2)")

if(bin6_7_8 >= 0.03 and bin6_7_8 < 0.1) :
    print("class 0(bin6_7_8)")

if(bin1_2 > 0.71 and bin1_2 <= 1.0) :
    print("class 2(bin1_2)")

if(bin6_7_8 == 0) :
    print("class 2(bin6_7_8)")

if(bin1_2 >= 0.44 and bin1_2 <= 0.57) :
    print("class 4(bin1_2)")

if(bin6_7_8 >= 0.1) :
    print("class 4(bin6_7_8)")


# frequency Block
if(mpfns >= 85 and mpfns < 95) :
    print("class 0(mpfns)")

if(mpfns >= 5 and mpfns <= 23) :
    print("class 2(mpnfs)")

if(mpfns >= 46 and mpfns <= 55) :
    print("class 4(mpfns)")
    

# Heart rate block
if(rns >= 23 and rns <= 37) :
    print("class 0(rns)")
    
if(qrsampns >= 0.21 and qrsampns <= 0.57) :
    print("class 2(qrsampns)")

if(heart_rate_block >= 6.172936 and heart_rate_block <= 6.2308960 ) :
    print("class 3(hrb)")
```
